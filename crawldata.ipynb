```python
import numpy as np
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.action_chains import ActionChains
import chromedriver_autoinstaller
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup
import time
from cgitb import text
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from asyncio.windows_events import NULL
from selenium.common.exceptions import TimeoutException
import csv
from this import d
import time
```

```python
url = 'https://vnexpress.net/suc-khoe'
url_page = '-p'
page = 20
link_href=[]

chromedriver_autoinstaller.install()
driver = webdriver.Chrome()
```

```python
for i in range(1,(page+1)):
    driver.get(url+url_page+str(i))
    time.sleep(5)
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight)")
    time.sleep(5)
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    #   tin dau tien
    news1 = soup.find('article',class_ ='item-news full-thumb article-topstory').find(class_ = 'title-news').find('a')
    link_href.append(news1.get('href'))
    # 3 tin tiep theo
    links_1 = soup.find_all('article',class_ = 'article-item')
    for link_news in links_1:
        links = link_news.find('a')
        link_href.append(links.get('href'))
    #tin con lai
    links_2 = soup.find_all('article',class_ = 'item-news item-news-common thumb-left')
    for link_news in links_2:
        if ("eclick_ad_holder ad_frame_protection" in str(link_news)):
            continue
        links = link_news.find(class_ = 'thumb-art').find(class_= 'thumb thumb-5x3')
        link_href.append(links.get('href'))
    link_href
```

```python
file = open('Link_suckhoe.txt','a')
for i in link_href:
    file.writelines(i +'\n')
file.flush()
file.close
```
